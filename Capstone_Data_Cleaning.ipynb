```python
# =============================================
# DATA CLEANING NOTEBOOK
# Project: AI-Driven OPD No-Show Analysis
# Dataset: Medical Appointment No Shows (Kaggle)
# =============================================

# Step 0: Import Libraries
import pandas as pd
import numpy as np

# Step 1: Load Dataset
# Upload your CSV file to Colab before running
from google.colab import files
uploaded = files.upload()

# Replace filename if needed
df = pd.read_csv('KaggleV2-May-2016.csv.zip', compression='zip')

print("\nInitial Data Preview:")
df.head()

# Step 2: Basic Data Inspection
print("\nData Info:")
df.info()

print("\nStatistical Summary:")
df.describe()

# Step 3: Drop Irrelevant Columns
# These do not contribute to prediction
cols_to_drop = ['PatientId', 'AppointmentID']
df.drop(columns=cols_to_drop, inplace=True)

# Step 4: Rename Columns for Consistency
df.rename(columns={
    'Hipertension': 'Hypertension',
    'Handcap': 'Handicap',
    'No-show': 'No_show'
}, inplace=True)

# Step 5: Convert Date Columns to Datetime
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

# Step 6: Create Waiting Days Feature (IMPORTANT)
df['WaitingDays'] = (df['AppointmentDay'] - df['ScheduledDay']).dt.days

# Step 7: Remove Invalid Waiting Days (Negative Values)
df = df[df['WaitingDays'] >= 0]

# Step 8: Clean Age Column
# Remove negative and unrealistic ages
df = df[df['Age'] >= 0]
df = df[df['Age'] <= 100]

# Step 9: Convert Target Variable (No_show)
df['No_show'] = df['No_show'].map({'Yes': 1, 'No': 0})

# Step 10: Encode Gender
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})

# Step 11: Create Additional Useful Features

# Day of Week Feature
df['AppointmentDayOfWeek'] = df['AppointmentDay'].dt.day_name()

# Age Group Feature
df['AgeGroup'] = pd.cut(df['Age'],
                        bins=[0, 12, 19, 35, 60, 100],
                        labels=['Child', 'Teen', 'Young', 'Adult', 'Senior'])

# Step 12: Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# Step 13: Remove Duplicates
df.drop_duplicates(inplace=True)

# Step 14: Optional - One Hot Encoding (if needed later)
# df = pd.get_dummies(df, columns=['Neighbourhood', 'AppointmentDayOfWeek', 'AgeGroup'], drop_first=True)

# Step 15: Drop Original Date Columns (Optional)
df.drop(columns=['ScheduledDay', 'AppointmentDay'], inplace=True)

# Step 16: Final Data Check
print("\nFinal Data Info:")
df.info()

print("\nFinal Preview:")
df.head()

# Step 17: Save Cleaned Dataset
cleaned_file_name = "cleaned_no_show_data.csv"
df.to_csv(cleaned_file_name, index=False)

# Download cleaned file
files.download(cleaned_file_name)

print("\nData cleaning completed successfully!")
```
